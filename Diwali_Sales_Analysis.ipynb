# Gender

```python
df["Gender"].value_counts()
# they are two types of genders F and M 
```

```python
ax=sns.countplot(x ="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater then men

# Age

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(["Age Group"],as_index=False,)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=sales_age)
```

From the Graphs we can see that most of the buyers are age group 26_35 yrs female

```python
df.columns
```

# State

```python
pd.DataFrame(df["State"]).value_counts().reset_index()
```

They are total 16 states are there 

```python
sales_state=df.groupby(["State"],as_index=False,)["Orders"].sum().sort_values("Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(x="State",y="Orders",data=sales_state)
```

```python
sales_amount=df.groupby(["State"],as_index=False,)["Amount"].sum().sort_values("Amount",ascending=False).head(10)
```

```python
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(x="State",y="Amount",data=sales_amount)
```

```python
df.columns
```

# Marital_Status

```python
df["Marital_Status"].value_counts()
```

They are 2 types of marital status that the represent the 0,1

```python
ax=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(6,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Marital_Status','Gender'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={"figure.figsize":(6,5)})
sns.barplot(data=sales,x="Marital_Status",y="Amount",hue="Gender")
```

From above Graph we can see that most of the buyers are married (women) and they have high purchasing power

```python
df.columns
```

# Occupation

```python
pd.DataFrame(df["Occupation"].value_counts()).reset_index()
```

They are 15 types of occupation are there

```python
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.countplot(data=df,x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupa=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(data=sales_occupa,x="Occupation",y="Amount")
```

From above the graphs we can see that most of the buyers are working in IT sector and Healthcare sectors

```python
df.columns
```

# Product_Category

```python
pd.DataFrame(df["Product_Category"].value_counts()).reset_index()
```

They 18 different products are there

```python
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.countplot(data=df,x="Product_Category")
plt.xticks(rotation=45)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.barplot(data=sales_product,x="Product_Category",y="Amount")
plt.xticks(rotation=45)
```

From above the graph we can see that most of the sold products are from food,Clothing,and Electronics

```python
df.columns
```

```python
sales_product_id = df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sales_product_id
```

```python
plt.figure(figsize=(10, 6))
sales_product_id.plot(x='Product_ID', y='Orders', kind='bar', legend=False)
plt.xticks(rotation=45)
```
